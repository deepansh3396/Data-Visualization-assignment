```python
#5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set a random seed for reproducibility
np.random.seed(42)

# Generate a synthetic dataset with correlated features
n_samples = 100
x = np.random.rand(n_samples)
y = 2 * x + np.random.normal(0, 0.1, n_samples)  # y is correlated with x
z = 0.5 * x + np.random.normal(0, 0.1, n_samples)  # z is also correlated with x

# Create a DataFrame
data = pd.DataFrame({
    'Feature X': x,
    'Feature Y': y,
    'Feature Z': z
})

# Calculate the correlation matrix
correlation_matrix = data.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar=True, linewidths=.5)

# Add title
plt.title('Correlation Matrix Heatmap', fontsize=16)

# Show the plot
plt.show()
```

```python
#3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and days

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set random seed for reproducibility
np.random.seed(20)

# Generate sales data
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

# Create a pivot table
pivot_table = df.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='mean', fill_value=0)

# Set the aesthetic style of the plots
sns.set(style="whitegrid")

# Create the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, annot=True, fmt=".1f", cmap="YlGnBu", linewidths=.5)
plt.title('Sales Variation Across Different Months and Days')
plt.xlabel('Month')
plt.ylabel('Day')
plt.show()
```

```python
#4. Using the given x and y data, generate a 3D surface plot to visualize the function



import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate x and y data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Create a DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Set up the figure and 3D axes
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Create the surface plot
surface = ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')

# Add a color bar which maps values to colors
fig.colorbar(surface, ax=ax, shrink=0.5, aspect=10)

# Set labels
ax.set_title('3D Surface Plot of z = sin(sqrt(x^2 + y^2))')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

# Show the plot
plt.show()
```

```python
#2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColumnDataSource

# Output in a Jupyter notebook
output_notebook()

# Generate random data
np.random.seed(42)  # For reproducibility
n = 100  # Number of points
x = np.random.rand(n) * 100  # Random x values
y = np.random.rand(n) * 100  # Random y values
sizes = np.random.randint(5, 30, size=n)  # Random sizes for markers
colors = np.random.rand(n)  # Random colors for markers

# Create a DataFrame for easier handling
data = pd.DataFrame({'x': x, 'y': y, 'sizes': sizes, 'colors': colors})

# Create a ColumnDataSource
source = ColumnDataSource(data=data)

# Create a Bokeh scatter plot
p = figure(title="Random Scatter Plot", x_axis_label='X', y_axis_label='Y', width=800, height=400)
# Added low and high arguments to linear_cmap, using the min and max values of the 'colors' column
p.scatter('x', 'y', size='sizes', color=linear_cmap('colors', 'Viridis256', low=min(data['colors']), high=max(data['colors'])), source=source, legend_field='sizes')

# Show the plot
show(p)
```

```python
#4. Create a Bokeh histogram to visualize the distribution of the given data. data_hist = np.random.randn(1000) hist, edges = np.histogram(data_hist, bins=30).

!pip install bokeh
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Output in a notebook (for Google Colab)
output_notebook()

# Generate random data
data_hist = np.random.randn(1000)

# Create histogram data
hist, edges = np.histogram(data_hist, bins=30)

# Prepare data for Bokeh
source = ColumnDataSource(data=dict(top=hist, left=edges[:-1], right=edges[1:]))

# Create a Bokeh histogram
p = figure(title="Histogram of Random Data", x_axis_label='Value', y_axis_label='Frequency', height=400, width=600)
p.quad(top='top', bottom=0, left='left', right='right', source=source, fill_color="navy", line_color="white")

# Show the plot
show(p)
```

```python
#5. Create a Bokeh heatmap using the provided dataset. data_heatmap = np.random.rand(10, 10) x = np.linspace(0, 1, 10) y = np.linspace(0, 1, 10) xx, yy = np.meshgrid(x, y)

!pip install bokeh
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar, ColumnDataSource
from bokeh.palettes import Viridis256

# Output in a notebook (for Google Colab)
output_notebook()

# Generate data for the heatmap
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Prepare data for Bokeh
data_flat = data_heatmap.flatten()
x_flat = xx.flatten()
y_flat = yy.flatten()

source = ColumnDataSource(data=dict(x=x_flat, y=y_flat, value=data_flat))

# Create a Bokeh heatmap
p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y',
           x_range=(0, 1), y_range=(0, 1),
           width=600, height=600)

# Create the heatmap
mapper = linear_cmap('value', palette=Viridis256, low=min(data_flat), high=max(data_flat))

p.rect(x='x', y='y', width=0.1, height=0.1, source=source,
        fill_color=mapper, line_color=None)

# Add a color bar
color_bar = ColorBar(color_mapper=mapper['transform'], location=(0,0))
p.add_layout(color_bar, 'right')

# Show the plot
show(p)
```
